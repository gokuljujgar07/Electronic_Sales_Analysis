I replaced the SKU codes with real product names in the 'Product' column. The codes were hard to understand, so I used a mapping based on US market products to make the data clear and readable.

```python

sku_to_product = {
    "SMP234": "Apple iPhone 14",
    "TBL345": "Samsung Galaxy Tab S8",
    "SKU1003": "Lenovo Yoga 7i",
    "SKU1004": "Google Pixel 7",
    "LTP123": "Dell XPS 13",
    "SWT567": "Apple Watch Series 8",
    "SKU1005": "HP Spectre x360",
    "HDP456": "Sony WH-1000XM5 Headphones",
    "SKU1002": "Amazon Fire HD 10",
    "SKU1001": "Samsung Galaxy S23"
}


Electronic_sales["Product"] = Electronic_sales["SKU"].replace(sku_to_product)
```

<div style="background-color: #8E8A92; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #FFFFFF;">Total Sales by Months</strong></h1>
</div>

```python
Electronic_sales["Months"] = Electronic_sales["Purchase Date"].dt.month_name()
```

```python
monthly_sales =  Electronic_sales.groupby('Months')['Revenue'].sum()
monthly_sales.div(1000000) # Convert to millions
```

```python
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
# Using 'month_order' as the list of possible categories (all months in calendar order)
# This tells pandas that these are the only valid categories, and their order matters
monthly_sales.index = pd.Categorical(monthly_sales.index, categories=month_order)
```

```python
 # 4 Total Sales by months
plt.figure(figsize=(10, 6))
sns.lineplot(x = monthly_sales.index, y = monthly_sales.values, marker="o",color = "orange")
plt.title("Total Sales by months", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales  (in millions)")
plt.grid(linestyle="--")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<span style="font-size:20px; font-weight:normal;">
    
Peak Sales Months: August, May, January, and June tied as highest, each close to **$6.9M**.
Stable Sales: January to August show consistent strong performance

 Decline Starts in September: Sales drop significantly in September, followed by a steep fall in October, November, and December.

 ðŸ“‰ Reason for Sales Drop in Quarter 4 
 
In the U.S., Quarter 4 (Octoberâ€“December) is usually the strongest sales period due to major events like Black Friday, Cyber Monday, and Christmas.

However, in this case, sales dropped sharply especially in December, which saw a 70% decline.

ðŸ” Possible Reasons:

- No holiday promotions or discounts

- Weak marketing or advertising after August

- Inventory shortages

- Poor checkout experience or delivery issues

- Business-specific seasonality or planning gaps

Customers expect deals during Quarter 4  without offers, they may shop elsewhere.
A review of customer behavior, inventory, and campaign planning is recommended to avoid similar drops in the future.


</span>

 June recorded the highest average order value, followed closely by August and September.

 This suggests that customers were more willing to make high-value purchases during mid-year.

<div style="background-color: #8E8A92; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #FFFFFF;">Age Grouping Function Based on Age Ranges</strong></h1>
</div>

```python
def age_group(age):
    if 18 <= age <= 30:
        return '18-30'
    elif 31 <= age <= 45:  
        return '31-45'
    elif 46 <= age <= 60:
        return '46-60'
    elif 61 <= age <= 80:
        return '61-80'
    else:
        return 'Out of Range'
```

```python
Electronic_sales['Age Group'] = Electronic_sales['Age'].apply(age_group)
```

```python
Electronic_sales[['Age','Age Group']].head()
```

```python
# 8. Sales by Age Group
plt.figure(figsize=(10, 6))
sns.histplot(data=Electronic_sales, x='Age Group', kde=True, color = "#FF0B55" )
plt.title('Sales Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```

<span style="font-size:20px; font-weight:normal;">
    
### ðŸ‘¥ Highest Sales by Age Group
 The 61â€“80 age group accounts for the highest sales frequency.
 
 Action: Continue investing in senior-friendly products, simplified ordering, and targeted loyalty programs.

### âœ… Target Age Group 18â€“35
Launch dedicated marketing campaigns on social platforms like Instagram, YouTube  to engage younger audiences.
For the 35â€“60 age group, personalize marketing efforts by segmenting offers based on lifestyle and preferences to boost engagement
