<div style="background-color: #8E8A92; padding: 25px; border-radius: 10px; font-size: 20px; font-weight: 10px; text-align: center;">
  <h1 style="color: #FFFFFF;">Age Grouping Function Based on Age Ranges</strong></h1>
</div>

```python
def age_group(age):
    if 18 <= age <= 30:
        return '18-30'
    elif 31 <= age <= 45:  
        return '31-45'
    elif 46 <= age <= 60:
        return '46-60'
    elif 61 <= age <= 80:
        return '61-80'
    else:
        return 'Out of Range'
```

```python
Electronic_sales['Age Group'] = Electronic_sales['Age'].apply(age_group)
```

```python
Electronic_sales[['Age','Age Group']].head()
```

```python
# 8. Sales by Age Group
plt.figure(figsize=(10, 6))
sns.histplot(data=Electronic_sales, x='Age Group', kde=True, color = "#FF0B55" )
plt.title('Sales Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.show()
```
